```python
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import OrdinalEncoder
import pandas as pd
import numpy as np
```

```python
data = {"Name": ["Aman","Riya","Rahul","Sneha","Arjun","Pooja","Karan","Neha","Vikram","Anjali"],
    "Gender": ["Male","Female","Male","Female","Male","Female","Male","Female","Male","Female"],
    "Department": ["CSE","IT","ECE","CSE","ME","IT","CSE","ECE","ME","CSE"],
    "City": ["Delhi","Mumbai","Kolkata","Delhi","Chennai","Mumbai","Delhi","Kolkata","Chennai","Delhi"],
    "Performance": ["Good","Average","Poor","Good","Average","Good","Good","Average","Poor","Good"]}
LE = LabelEncoder()
```

```python
df = pd.DataFrame(data)
```

```python

print("Dataset Created Successfully:\n")
print(df)
```

```python
LE = LabelEncoder()
df['Gender_encoded'] = LE.fit_transform(df['Gender'])
print(df)
```

```python
ohe = OneHotEncoder(sparse_output=False)
encoded_features = ohe.fit_transform(df[['Department']])
df = pd.DataFrame(encoded_features)
print(df)
```

```python
data = {
    "sname":['S1','S2','S3','S4'],
    "Height":[162,171,155,165],
    "Weight":[63,60,50,62],
    "Speed":[5,3,7,2]
}
```

```python
dataset = pd.DataFrame(data)
print(dataset)
```

```python
from sklearn.preprocessing import Normalizer
norm = Normalizer()
cols = dataset[["Height","Weight","Speed"]]
data = norm.fit_transform(cols)
df = pd.DataFrame(data,columns=cols.columns)

print(df)
```

```python
from sklearn.preprocessing import StandardScaler
SS = StandardScaler()
cols = dataset[["Height","Weight","Speed"]]
data = SS.fit_transform(cols)
df = pd.DataFrame(data,columns=cols.columns)

print(df)
```
